**Задание 1. Столбец Source_type**

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source == 'google'), 'source_type'] = 'organic'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region == 'Russia'), 'source_type'] = 'ad'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region != 'Russia'), 'source_type'] = 'other'
#если значение в столбце NaN, то заменяем его значением из столбца траффик_соурс
log['source_type'] = log['source_type']. fillna(log['traffic_source'])
log.head(15)
```

**Задание 3. Среднее время жизни пользователей**

```python
rating = pd.read_csv('ratings.csv')
#тут я сначала нахожу кол-во оценок у каждого пользователя, добавляю в новый датафрейм гроупед столбцы с минимальным, максимальным временем, а также с разницей (время жизни)
#для каждого пользователя, у которого количество оценок >100
grouped = rating.groupby('userId').count().reset_index()[['userId', 'rating']]
grouped['timestamp_max'] = rating.groupby('userId').max().reset_index()['timestamp']
grouped['timestamp_min'] = rating.groupby('userId').min().reset_index()['timestamp']
grouped['live_time'] = (grouped['timestamp_max']- grouped['timestamp_min'])[grouped['rating'] > 100]

grouped.head(30)
```

**Задание 4. Объединение таблиц**

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)

auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)

air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)

client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
```

```python
from functools import reduce
vehicle = reduce(lambda a,b: a.merge(b, how='outer', on='client_id'),[rzd, auto, air]) #сразу объединили 3 таблицы по столбцу клиент-ид
vehicle = vehicle.fillna(0)
vehicle
```

```python
full_info = vehicle.merge(client_base, how='outer', on='client_id') #объединили получившуюся таблицу с выручками с таблицей с адресами
full_info
```
