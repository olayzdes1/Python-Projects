Результат классификации запишите в столбец class.

**Задание 1. Классификация фильмов по рейтингу**

```python
import pandas as pd
```

```python
def rate_class(rating):
  if rating <= 2:
    return ('Низкий рейтинг')
  if 2 < rating <= 4:
    return ('Средний рейтинг')
  if rating == 4.5 or rating == 5:
    return ('Высокий рейтинг')
  else:
    return ('Ошибка')
```

```python
rating = pd.read_csv('ratings.csv')

rating['class'] = rating['rating'].apply(rate_class)
rating.head()
```

**Задание 2. Гео-классификатор**

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}

def geo_class(keyword):
  for region, cities in geo_data.items():
    for city in cities: #проходим по каждому городу региона
      if city in keyword: #если город есть в ключевом запросе
        return region
  return ('undefined')

keyword = pd.read_csv('keywords.csv')
keyword['region'] = keyword['keyword'].apply(geo_class)

keyword[keyword['region'] !='undefined'] #фильтр для вывода запросов только с городами
```

**Задание 3. Средний рейтинг фильмов**

```python
import numpy as np
```

```python
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')

years = np.arange(1950, 2011) #создали список с годами

def production_year(title):
  for year in years:
    if str(year) in title: #если год из списка years есть в названии фильма, то возвращ. этот год
      return year
  return 1900

movies['year'] = movies['title'].apply(production_year)
data = movies.merge(ratings, on='movieId')
data.groupby(['year']).mean(['rating']).sort_values(['rating'], ascending=False)

```
