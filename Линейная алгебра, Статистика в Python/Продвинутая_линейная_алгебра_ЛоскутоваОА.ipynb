**Задание 1.** Дан объект в 2D пространстве.
При помощи линейных отображений:
*   Уменьшить объект в два раза
*   Повернуть на 130 градусов
*   Отразить объект относительно прямой y=x

```python
import numpy as np
import matplotlib.pyplot as plt
import math

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)
transform_reduce = A @ np.array([[0.5,0,0], [0,0.5,0], [0,0,1]])
#делаем поворот на 130 градусов по часовой стрелке, поэтому аргумент -130 градусов
transform_rotate = A @ np.array([[math.cos(-130/180*math.pi), math.sin(-130/180*math.pi), 0], [-(math.sin(-130/180*math.pi)), math.cos(-130/180*math.pi), 0], [0,0,1]])
transform_mirror =  A @ np.array([[0,1,0], [1,0,0], [0,0,1]])

x = A[:,0]
y = A[:,1]

x1 = transform_reduce[:,0]
y1 = transform_reduce[:,1]

x2 = transform_rotate[:,0]
y2 = transform_rotate[:,1]

x3 = transform_mirror[:,0]
y3 = transform_mirror[:,1]

fig, ax = plt.subplots(2, 2)
fig.tight_layout(h_pad= 2)
ax[0,0].plot(x, y)
ax[0,0].set_ylim([-200, 200])
ax[0,0].set_xlim([-200, 200])
ax[0,0].set_title('Начальный объект')

ax[0,1].plot(x1, y1)
ax[0,1].set_ylim([-200, 200])
ax[0,1].set_xlim([-200, 200])
ax[0,1].set_title('Уменьшение в 2 раза')

ax[1,0].plot(x2, y2)
ax[1,0].set_ylim([-200, 200])
ax[1,0].set_xlim([-200, 200])
ax[1,0].set_title('Поворот на 130 градусов')

ax[1,1].plot(x3, y3)
ax[1,1].set_ylim([-200, 200])
ax[1,1].set_xlim([-200, 200])
ax[1,1].set_title('Отражение относительно у=х')

plt.show()
```

**Задание 2.** Найдите спектральное разложение матрицы. Формула A = S*D*St

```python
m = np.array([[1, 2],
              [2, 3]])

D = np.diag(np.linalg.eigvals(m))
S = np.linalg.eig(m)[1]
St = np.transpose(S)

print(f'Исходная матрица m:\n{m}')
print(f'Компоненты спектрального разложения:\nМатрица c собственными векторами по столбцам S:\n{S},\n Диагональная матрица D с собственными значениями по главной диагонали:\n{D},\nТранспонированная матрица St:\n{St}')
print(f'Проверка на корректность разложения:\n {np.dot(S, np.dot(D,St))}')
```
