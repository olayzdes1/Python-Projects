Задание 1. В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта

```python
P_A1 = 0.2 #вероятность того, что продукция первого предприятия
P_A2 = 0.3
P_A3 = 0.5
P_B_A1 = 0.1 #вероятность того, что продукция первого предприятия высшего сорта
P_B_A2 = 0.05
P_B_A3 = 0.2
P_B = P_A1*P_B_A1 + P_A2*P_B_A2 + P_A3*P_B_A3
print(f'Вероятность того, что случайно купленная новая продукция высшего сорта: P(B) = {P_B}')
```

Задание 3. Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

Параметры для построения графика плотности нормального распределения — loc (среднее значение) и scale (дисперсия)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, t
```

```python
x = np.linspace(-6, 6, 100)
y1 = norm.pdf(x, loc=0, scale=1)
y2 = norm.pdf(x, loc=2, scale=1)
y3 = norm.pdf(x, loc=0, scale=3)
y4 = norm.pdf(x, loc=1, scale=2)
y5 = norm.pdf(x, loc=-1, scale=0.5)
```

```python
plt.plot(x, y1, label='Норм. распред.1')
plt.plot(x, y2, label='Норм. распред.2')
plt.plot(x, y3, label='Норм. распред.3')
plt.plot(x, y4, label='Норм. распред.4')
plt.plot(x, y5, label='Норм. распред.5')
plt.xlabel('x')
plt.ylabel('Плотность')
plt.title('График плотности нормального распределения')
plt.legend()
plt.show()
```

На графике видно, что при увеличении значения loc график плотности нормального распределения сдвигается вправо (распред. 2); при увеличении значения scale график становится шире (распред. 3); при увеличении loc и scale график сдвигается вправо и становится шире (распред. 4); при уменьшении scale и loc график становится уже и центр сдвигается влево.

Параметр для построения графика плотности экспоненциального распределения — scale (параметр масштаба, scale>0)

```python
x = np.linspace(0, 5, 100)
y1 = expon.pdf(x, scale=1)
y2 = expon.pdf(x, scale=0.5)
y3 = expon.pdf(x, scale=0.7)
y4 = expon.pdf(x, scale=1.5)
y5 = expon.pdf(x, scale=2)
```

```python
plt.plot(x, y1, label='Экспон. распред.1')
plt.plot(x, y2, label='Экспон. распред.2')
plt.plot(x, y3, label='Экспон. распред.3')
plt.plot(x, y4, label='Экспон. распред.4')
plt.plot(x, y5, label='Экспон. распред.5')
plt.xlabel('x')
plt.ylabel('Плотность')
plt.title('График плотности экспоненциального распределения')
plt.legend()
plt.show()
```

По графику можно сделать вывод о том, что чем меньше значение scale, тем больше значение плотности и тем сильнее перегиб; а при увеличении значения scale график становится более пологим, больше напоминающим линейную зависимость  

Параметры для построения графика плотности распределения Стьюдента — loc, scale и df (количество степеней свободы)

```python
x = np.linspace(-7, 7, 100)
y1 = t.pdf(x, df=5, loc=0, scale=1)
y2 = t.pdf(x, df=2, loc=-2, scale=1)
y3 = t.pdf(x, df=5, loc=0, scale=2)
y4 = t.pdf(x, df=7, loc=3, scale=3)
y5 = t.pdf(x, df=0.5, loc=-1, scale=0.7)
```

```python
plt.plot(x, y1, label='Распред. Стьюдента 1')
plt.plot(x, y2, label='Распред. Стьюдента 2')
plt.plot(x, y3, label='Распред. Стьюдента 3')
plt.plot(x, y4, label='Распред. Стьюдента 4')
plt.plot(x, y5, label='Распред. Стьюдента 5')
plt.xlabel('x')
plt.ylabel('Плотность')
plt.title('График плотности экспоненциального распределения')
plt.legend()
plt.show()
```

Параметры loc и scale влияют на график также как и в графике плотности нормального распределения; уменьшение параметра df приводит к более острому перегибу распределения
